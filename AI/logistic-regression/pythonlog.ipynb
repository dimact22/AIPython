```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
n = 300
ages = np.random.randint(18, 60, size=n)
salaries = np.random.randint(20000, 100000, size=n)
bought = (ages > 30) & (salaries > 50000)
bought = bought.astype(int)


df = pd.DataFrame({
    'Age': ages,
    'Salary': salaries,
    'Bought': bought
})

print(df.head())
```

```python
X = df[['Age', 'Salary']]
y = df['Bought']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Salary', hue='Bought', palette='coolwarm')
plt.title("Age vs Salary with Purchase Decision")
plt.xlabel("Age")
plt.ylabel("Salart")
plt.show()
```
